```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc

rc('text', usetex=False)
import matplotlib
matplotlib.rcParams['font.sans-serif'] = "Arial"
matplotlib.rcParams['font.family'] = "sans-serif"
```

```python
T = 100

xi =  7
lamb = 1

dt = lamb/25
num = int(T/dt)

x = np.zeros(num)
y = np.zeros(num)
theta = np.zeros(num)
K = np.zeros(num)

D = 1/(lamb*xi*xi)
kappa = 0.05

K[0] = np.random.randn()*np.sqrt(D*lamb)

for i in range(num-1):
    x[i+1] = x[i] + dt*np.cos(theta[i])
    y[i+1] = y[i] + dt*np.sin(theta[i])
    theta[i+1] = theta[i] + dt*K[i] + np.sqrt(kappa*dt)*np.random.randn()
    K[i+1] = K[i] - dt*K[i]/lamb + np.sqrt(D*dt)*np.random.randn()

plt.close("all")
fig,axis = plt.subplots(1,1,figsize = (10,10))
axis.spines['top'].set_linewidth(1.25)
axis.spines['left'].set_linewidth(1.25)
axis.spines['bottom'].set_linewidth(1.25)
axis.spines['right'].set_linewidth(1.25)
axis.tick_params(labelsize = 25)

axis.plot(x,y,'k-',lw=3)
axis.set_aspect('equal')
# plt.axis('off')

fig.tight_layout()
plt.show()
```
